```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
aseg_stats = pd.read_csv("COMPARISON/comparison_aseg.stats.csv")

aseg_stats = aseg_stats.dropna().sort_values("Structure", ascending = True)
aseg_stats_logP = pd.DataFrame(aseg_stats["(-)log10(p)"])

aseg_stats_label = list(aseg_stats["Structure"].values)

plt.rcParams["figure.figsize"] = (1,25)
sns.heatmap(aseg_stats_logP, yticklabels=aseg_stats_label,
            annot=True,
            fmt=".2f",
            linewidths=1,
            linecolor='red',
            annot_kws={'rotation': 0},
            cbar_kws={"shrink": 10, "pad" : 0.2})
plt.savefig("HEATMAP/INDEX/HeatMap_aseg.stats.png",dpi=300, bbox_inches = "tight")
plt.show()
```
